7) Write a Python program to combine two dictionary adding values for common keys.

d1 = {'a': 100, 'b': 200, 'c':300}

d2 = {'a': 300, 'b': 200, 'd':400}

Sample output:

`Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})`

````python
from collections import Counter

d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}

mydictionary=Counter(d1) + Counter(d2)
print(mydictionary)
````

13) Write a Python program to match key values in two dictionaries.

Sample dictionary: {'key1': 1, 'key2': 3, 'key3': 2}, {'key1': 1, 'key2': 2}

Expected output:

`key1: 1 is present in both x and y`

````python
x={'key1': 1, 'key2': 3, 'key3': 2}
y={'key1': 1, 'key2': 2}

common={key:value for key,value in x.items() if key in y and y[key]==value}
for key, value in common.items():
  print(f"{key}:{value} is present in both x and y")
````

14) # Write a Python program to print a dictionary line by line.

Players={'Rohit':{'runs':10000,'centuries':15},'Virat':{'runs':12000,'centuries':18}}

expected output:
```
Rohit
runs :	 10000
centuries :	 15
Virat
runs :	 12000
centuries :	 18
```

````python
Players={'Rohit':{'runs':10000,'centuries':15},'Virat':{'runs':12000,'centuries':18}}

for key,value in Players.items():
  print(key)
  for key, value in value.items():
    print(f"{key}:\t{value}")
````

16) Write a Python program to remove spaces from dictionary keys.

Students={'d 01':'Abc','d 0 2':'Xyz','d0 3':'Pqr'}

`{'d01': 'Abc', 'd02': 'Xyz', 'd03': 'Pqr'}`


19) Write a Python script to concatenate following dictionaries to create a new one.

Sample Dictionary :

dic1={1:10, 2:20}

dic2={3:30, 4:40}

dic3={5:50,6:60}

Expected Result : `{1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}`

````python
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}

mydictionary={key:value for i in [dic1,dic2,dic3] for key,value in i.items()}
print(mydictionary)
````
